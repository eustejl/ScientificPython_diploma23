### **Exercise2**

Write a decorator that plots a function (of one variable) over the interval (0,100). Make sure your plot has a name, the axis names, nice color, changed background color and a legend.

Do the same using a different "interface".

```python
def plot_decorator(f):
    def wrapper(*args,**kwargs):
        x = np.linspace(0,100)
        plt.plot(x,f(x))
        plt.show()
    return wrapper
        
        
```

```python
@plot_decorator
def sine(x):
    return np.sin(x*np.pi/180)
```

```python
print(sine(5))
```

### **Exercise3:**

Plot the recaman sequence in the following way:

Add a semi-circular arc from $a_{n}$ to $a_{n-1}$.

Alternate the arcs to be above and below the x-axis according to whether
    n is even or odd.

```python
def recaman(n):
    rec_list = [0]
    
    for i in range(1,n):
        a_nm = rec_list[i-1] - i
        if (a_nm > 0) and not(a_nm in rec_list[:i]):
            rec_list.append(a_nm)
        else:
            rec_list.append(rec_list[i-1] + i)
    
    return rec_list 
```

```python
reca = recaman(30)
```

```python
fig,ax = plt.subplots()

for i in range(len(reca)-1):
    a=reca[i]; b=reca[i+1];
    c = (a+b)/2.
    r = abs(c-a)
    
    X = np.linspace(a,b)
    Y = (r**2 - (X-c)**2)**1/2

    ax.axhline(0,color='k',lw=0.1)
    if (i+1)%2==0: ax.plot(X,Y,'b')
    else: ax.plot(X,-Y,'r')
plt.show()
```

### **Exercise4:**

Plot a rainbow.

```python
w = 0.01
colors = ['red','orange','yellow','green','blue','indigo','violet']

fig,ax = plt.subplots()
for i in range(7):
    a,b = -(w*i+1),w*i+1
    c = (a+b)/2
    r = abs(c-a)
    X = np.linspace(a,b)
    Y = (r**2 - (X-c)**2)**1/2

    ax.plot(X,Y,color=colors[i],lw=3)

plt.show()
```
